```python
import matplotlib.pyplot as rajamouli
movie=['stu.no.1','yamadonga','maryada ramana','eega','rrr']
budget=[12,40,30,125,1200]
rajamouli.pie(budget,labels=movie)
ntr=rajamouli.Circle((0,0),0.6,color='white')
rajamouli.gca().add_artist(ntr)
rajamouli.show()
```

```python
import matplotlib.pyplot as garu
in2013=[7,3,8]
in2014=[4,2,10]
in2015=[6,8,4]
actress=['trisha','kajal','anushka']
data=[in2013,in2014,in2015]
garu.boxplot(data,labels=actress)
garu.show()
```
